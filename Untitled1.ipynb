```python
cars=pd.read_csv('mtcars.csv')
cars.head()
```

```python
plt.hist(cars['mpg'],rwidth=0.9)
plt.title('Frequency distribution of MPG')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()

interval=cars['mpg'].value_counts().idxmax()
interval
```

```python
plt.boxplot(cars['mpg'])
plt.show()
print("The five number summary:")
cars['mpg'].describe()
```

## 9

```python
Series_A=pd.Series([10, 20, 30, 40, 50])
Series_B =pd.Series([40, 50, 60, 70, 80])
print(Series_A)
print(Series_B)
```

```python
u=pd.Series(np.union1d(Series_A,Series_B))
i=pd.Series(np.intersect1d(Series_A,Series_B))
not_com=u[~u.isin(i)]
not_com
```

```python
maxA=Series_A.max()
minA=Series_A.min()
print(f"The maximum value={maxA}")
print(f"The minimum value={minA}")
```

```python
sumB=Series_B.sum()
print(f"The sum of Series_B= {sumB}")
```

```python
midB=Series_B.median()
print(f"The median of Series_B= {midB}")
```
